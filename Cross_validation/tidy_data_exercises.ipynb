1. Attendance Data

Load the attendance.csv file and calculate an attendnace percentage for each student. One half day is worth 50% of a full day, and 10 tardies is equal to one absence.

You should end up with this:


| student | score  |
| ------- | -----  |
| Billy   | 0.2250 |
| Jane    | 0.1875 |
| John    | 0.7125 |
| Sally   | 0.3625 |

```python
attendance = pd.read_csv('untidy_data/attendance.csv')
attendance
```

```python
attendance.rename(index=str, columns={'Unnamed: 0': 'name'}, inplace=True)
```

```python
# melts one set of variables in columns into rows for each id_vars variable
# keep id_vars, change var_name (default variable) and put values into value_name (default value)
melted = attendance.melt(id_vars=['name'], var_name='date', value_name='response')
melted.head()
```

```python
response_types = {'P': 1, 'H': .5, 'T': .9, 'A': 0}
melted['score'] = melted['response'].map(response_types)
melted.head()
```

```python
melted.groupby('name').mean()
```

3. Cake Recipes

    a. Read the cake_recipes.csv data. This data set contains cake tastiness scores for combinations of different recipes, oven rack positions, and oven temperatures.

    b. Tidy the data as necessary.

    c. Which recipe, on average, is the best? recipe b
    
    d. Which oven temperature, on average, produces the best results? 275
    
    e. Which combination of recipe, rack position, and temperature gives the best result? recipe b, bottom rack, 300 degrees

```python
recipes = pd.read_csv('untidy_data/cake_recipes.csv')
recipes
```

```python
cakes = recipes.melt(id_vars=['recipe:position'], var_name='temp', value_name='tastiness')
```

```python
cakes.head()
```

```python
cols = cakes['recipe:position'].str.split(':', expand=True)
cols.columns = 'recipe', 'position'
cols.head()
```

```python
cakes = pd.concat([cakes.drop(columns='recipe:position'), cols], axis=1)
```

```python
cakes.groupby('recipe').tastiness.mean()
```

```python
cakes.sort_values(by=['temp','tastiness'], ascending=True).tail(1)
```
